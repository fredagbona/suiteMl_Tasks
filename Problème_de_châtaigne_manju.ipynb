[Problème 1] Création de nombres aléatoires

```python
import numpy as np

np.random.seed(0) # Fixing random state for reproducibility

mean = [-3, 0]
cov = [[1., .8], [.8, 1.]]

a_list = np.random.multivariate_normal(mean, cov, (500))

print(a_list.shape)
```

[Problème 2] Visualisation par nuage de points

```python
import matplotlib.pyplot as plt


x = a_list[:,0] # extract first column
y = a_list[:,1] # extract second column

plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(y,x,c="b",alpha=0.5,label="factice grains")

plt.show()
```

[Problème 3] Visualisation par histogramme

```python
plt.title("histogram x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(x,50, density=True, facecolor='b', alpha=1) 
plt.xlim(-5, 2)

plt.show()
```

```python
plt.title("histogram x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(y,50, density=True, facecolor='r', alpha=1)
plt.xlim(-5, 2)
plt.show()
```
